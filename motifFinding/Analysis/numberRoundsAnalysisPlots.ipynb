```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
data_ic15=[]
data_ic1=[]
data_ic2=[]

f_rounds = open('rounds','rb')

for line in f_rounds:
    cols=line.strip('\n').split(' ')
    if cols[0][:11] == 'dataset_1.5':
        data_ic15.append(int(cols[1]))
    if cols[0][:10] == 'dataset_1_':
        data_ic1.append(int(cols[1]))  
    if cols[0][:19] == 'dataset_2_8_500_10_':
        data_ic2.append(int(cols[1]))
        
f_rounds.close()

avg1 = float(sum(data_ic1))/len(data_ic1)
avg15 = float(sum(data_ic15))/len(data_ic15)
avg2 = float(sum(data_ic2))/len(data_ic2)
```

```python
data_ml6=[]
data_ml7=[]
data_ml8=[]

f_rounds = open('rounds','rb')

for line in f_rounds:
    cols=line.strip('\n').split(' ')
    if cols[0][:12] == 'dataset_2_6_':
        data_ml6.append(int(cols[1]))
    if cols[0][:12] == 'dataset_2_7_':
        data_ml7.append(int(cols[1]))  
    if cols[0][:19] == 'dataset_2_8_500_10_':
        data_ml8.append(int(cols[1]))
        
f_rounds.close()

avg6 = float(sum(data_ml6))/len(data_ml6)
avg7 = float(sum(data_ml7))/len(data_ml7)
avg8 = float(sum(data_ml8))/len(data_ml8)
```

```python
data_sc5=[]
data_sc10=[]
data_sc20=[]

f_rounds = open('rounds','rb')

for line in f_rounds:
    cols=line.strip('\n').split(' ')
    if cols[0][:19] == 'dataset_2_8_500_20_':
        data_sc20.append(int(cols[1]))
    if cols[0][:18] == 'dataset_2_8_500_5_':
        data_sc5.append(int(cols[1]))  
    if cols[0][:19] == 'dataset_2_8_500_10_':
        data_sc10.append(int(cols[1]))
        
f_rounds.close()

avgsc5 = float(sum(data_sc5))/len(data_sc5)
avgsc10 = float(sum(data_sc10))/len(data_sc10)
avgsc20 = float(sum(data_sc20))/len(data_sc20)
```

```python
list_ic=[avg2,avg15,avg1]
list_ml = [avg6,avg7,avg8]
list_sc = [avgsc5,avgsc10,avgsc20]

plt.plot(list_ic,'D')
plt.plot(list_ic,'b',label='ICPC')

plt.plot(list_ml,'D')
plt.plot(list_ml,'g',label='ML')


plt.plot(list_sc,'D')
plt.plot(list_sc,'r',label='SC')

plt.ylabel('#Rounds on average')
plt.xlabel('Complexity of the problem')
plt.tick_params(axis='x',bottom='off',top='off',labelbottom='off')
plt.xlim(-0.25,2.25)
plt.ylim(10000,105000)
plt.legend(frameon=False,loc=(0,0.7))
plt.rc("savefig", dpi=900)
```
